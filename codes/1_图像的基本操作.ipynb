```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# 1. 读取图 cv2.imread
img = cv2.imread('../images/car.jpg', 0) # type: np.ndarray
```

```python
# 2. 显示图 cv2.imshow
cv2.imshow('image', img)
```

```python
# 利用matplotlib显示窗口
plt.imshow(img, cmap='gray', interpolation='bicubic'), plt.axis("off")
```

```python
# 3. 保存图 cv2.imwrite
cv2.imwrite("../images/car2.jpg", img)
```

```python
# 利用matplotlib保存图片
plt.savefig("../images/car3.png")
plt.show()
```
